```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import datetime as dt
import os
import numpy as np
```

```python
pwd = os.getcwd()

files = [file for file in os.listdir(pwd) if file.endswith(".csv")]

df_blank = pd.DataFrame()

for file in files:
    data = pd.read_csv(file)
    dataset =  pd.concat([df_blank, data])

dataset
```

```python
dataset.isnull().sum()
```

Let us also check for the number of unique values in our dataset.

```python
dataset.nunique()
```

### Data Cleaning

The coordinates of different stations are not necessary for our analysis. We can drop them.

```python
dataset = dataset.drop(columns=["start_lat", "start_lng","end_lat","start_lng","end_lng"])
dataset
```

There are still null values from the *station names* and *station ids*. This is maybe due to data collection by the system. We can drop these null values from the said columns.

Re-checking if there are still null values from the dataset.

```python
dataset.isnull().sum()
```

As we can see here, started_at and ended_at columns are in object data type format.
We need to convert this into datetime format for further analysis.

```python
dataset.info()
```

```python
# Changing the datatype using pd.to_datetime.

dataset["started_at"] = pd.to_datetime(dataset["started_at"], format = "%Y-%m-%d %H:%M:%S")
dataset["ended_at"] = pd.to_datetime(dataset["ended_at"], format = "%Y-%m-%d %H:%M:%S")
```

Let us change the name of column for membership.

```python
dataset.rename(columns={"member_casual":"membership_type"}, inplace=True)
dataset
```

### Data transformation

We need to know how much time for each ride. We can get that by subtracting the ended_at and started_at columns.

```python
# First we subtract the end time from the start time of each trip.
# Then we convert it in seconds using dt.total_seconds().
# Lastly we converted it into minutes and rounding off the values to two decimal places.


dataset["trip_duration_mins"] = round(((dataset["ended_at"] - dataset["started_at"]).dt.total_seconds()) / 60, 2)
dataset
```

Checking if there are zero or negative values on the trip duration column.

```python
dataset[(dataset["trip_duration_mins"] < 0) | (dataset["trip_duration_mins"] == 0)]
```

We need only the data with trip duration (mins) that is greater than 0.

```python
dataset = dataset[dataset["trip_duration_mins"] > 0]
dataset
```

For our analysis we need to know the days of the week of each ride. For this we can use the dt.day_name() function to our started_at column. This will return the days of the week for each ride.

```python
dataset["start_day"] = dataset["started_at"].dt.day_name()
dataset["end_day"] = dataset["ended_at"].dt.day_name()
dataset
```

We can start aggregating the data from here.

```python
# Finding the average, maximum, and minimum trip duration per day.

trip_duration_day = dataset.groupby("start_day").agg({"trip_duration_mins": ["mean", "max", "min"]}).reset_index()
trip_duration_day
```

Now let us find out the average, minimum, and maximum duration (minutes) by each membership type.

```python
trip_duration_membership = dataset.groupby("membership_type").agg({"trip_duration_mins": ["mean","max","min"]}).reset_index()
trip_duration_membership
```

Furthermore, let us find out now the average, minimum, and maximum trip duration by each membership type during the days of the week.

```python
membership_day = dataset.groupby(["membership_type", "start_day"]).agg({"trip_duration_mins": ["mean","max","min"]}).reset_index()
pivot1 = pd.pivot_table(membership_day, values="trip_duration_mins", index="membership_type",columns="start_day")
 # pivot1.to_csv("membership_descriptive_stats.csv", index=True)
pivot1
```

Let us find out the distribution of member and casual riders with the different types of bike they are using.

```python
sns.catplot(x="rideable_type",data=dataset,hue="membership_type",kind='count')
```

It is interesting to see that members used classic bikes the most. Here we can also see that no members uses docked bikes but rather casual members only uses this type of bike.

Let us now see the distribution of each membership type during the days of the week.

```python
sns.catplot(x="start_day",data=dataset,hue="membership_type",kind='count', aspect=4)
plt.xlabel("Days of the week", fontsize=15)
```

Here we can see that member riders uses cycling service more than casual riders during weekdays. On weekends, casual riders uses the service more than member riders.

We can add another column for hours to see when is the spike for each day where the number of users is the highest.

```python
dataset["hour"] = dataset["started_at"].dt.hour
dataset
```

```python
fig = sns.FacetGrid(dataset,hue='start_day',aspect=6)

fig.map(sns.kdeplot,'hour')


fig.set(xlim=(0,23), xticks=(range(0,24)))
fig.set_xlabels("Hour of the day", fontsize=15)

fig.add_legend()
```

It is interesting to note here that during weekdays, spikes for using the service happen during early in the morning (between 7-8 am) and late in the afternoon (between 5-6 pm).
On weekends, spikes for using the cyclistic service happens in noontime (12-1 pm). Maybe we can target to place the advertisements before these certain spike periods for effective
ad campaign. 

Let us see now where are the spikes during the day that each membership type uses the cyclistic service.

```python
fig = sns.FacetGrid(dataset,hue='membership_type',aspect=6)

fig.map(sns.kdeplot,'hour')

fig.set(xlim=(0,23), xticks=(range(0,24)))
fig.set_xlabels("Hour of the day", fontsize=15)

fig.add_legend()
```

It is interesting to note that members' usage of the cyclistic service spikes up early in the morning and late in the afternoon. On the other hand, casual riders use the service with an increasing trend from morning reaching the maximum during late afternoon (5-6 pm).
